```python
import pandas as pd
```

```python
torrents = pd.read_csv('torrents.csv')
```

```python
a = torrents.sort_values("totalSnatched",ascending= False).head()
print(a)
```

```python
tags = pd.read_csv('tags.csv')
```

```python
tags.head()
```

```python
filter = (torrents["groupYear"] < 2000)
all_tags_before2000 = []
data = torrents[filter].sort_values("totalSnatched",ascending= False).head(1000)
for item_id in data["id"]:
    filter = ((tags["id"]) == item_id)
    for item_tag in tags[filter]["tag"]:
        all_tags_before2000.append(item_tag)
print(len(all_tags_before2000))
print(all_tags_before2000)
```

```python
from collections import Counter
Counter(all_tags_before2000).most_common()
```
